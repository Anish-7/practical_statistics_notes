    Simple Linear Regression: (between two variables: one feature and one target)

- Simple linear regression models the relationship between to two variables

- It models how much does the variable target increases/decrease with respect to x feature

- It is similar to correlation which measures the strenght of association between two variables

- But simple linear regresssion quantifies(How much?) that relationships.

- ols - ordinary least squares(of error) 

    - It is method of fitting regression model by minimizing the the sum of squared residuals (RSS).

    - RSS = SUMMATION (Y-Y^)**2

Equation of simple linear regression:

- y = B1(X) + B0

- where B1 and B0 are the coefficient that has to be fitted

    - B0 is the intercept : it is value of y when x=0

```python
path ='/home/anish/plan/practical_statistics_notes/data/LungDisease.csv'

import pandas as pd
import numpy as np
import seaborn as sns

lung = pd.read_csv(path)

sns.scatterplot(x=lung.Exposure , y= lung.PEFR)

from sklearn.linear_model import LinearRegression

model=LinearRegression()

exp=np.array(lung.Exposure).reshape(-1,1)
pefr=np.array(lung.PEFR)
pefr
model.fit(exp,pefr)
b0=model.intercept_
b1=model.coef_

print(f'intercept: {b0} coefficent b1 : slope {b1} slope is negeative')
```
